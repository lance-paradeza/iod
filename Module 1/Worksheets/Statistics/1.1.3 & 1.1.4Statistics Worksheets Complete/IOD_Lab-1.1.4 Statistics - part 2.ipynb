````python
X = np.array([0, 5, 10, 15, 20])
Y = np.array([0, 7, 10, 13, 20])

def best_fit(X, Y):
    n = len(X)
    
    sum_x = np.sum(X)
    sum_y = np.sum(Y)
    sum_xy = np.sum(X * Y)
    sum_x_squared = np.sum(X**2)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
    b = (sum_y - m * sum_x) / n
    
    return m, b

m, b = best_fit(X, Y)
fit_line = m * X + b

plt.scatter(X, Y, label='Data Points')
plt.plot(X, fit_line, color='magenta', label=f'Best Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Create a single figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Add the first subplot (probability plot against normal distribution)
x = stats.loggamma.rvs(5, size=500) + 5
stats.probplot(x, dist='norm', plot=ax1)
ax1.set_xlabel('Theoretical Quantiles')
ax1.set_ylabel('Ordered Values')
ax1.set_title('Probability Plot against Normal Distribution')

# Add gridlines for the first subplot
ax1.grid(True, linestyle='--', alpha=0.7, color='lightgray')

# Add the second subplot (probability plot after Box-Cox transformation)
# Perform Box-Cox transformation
x_transformed, _ = stats.boxcox(x)

# Probability plot after Box-Cox transformation
stats.probplot(x_transformed, dist='norm', plot=ax2)
ax2.set_xlabel('Theoretical Quantiles')
ax2.set_ylabel('Ordered Values')
ax2.set_title('Probability Plot after Box-Cox Transformation')

# Add gridlines for the second subplot
ax2.grid(True, linestyle='--', alpha=0.7, color='lightgray')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````

````python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

data = [170, 167, 174, 179, 179, 187, 179, 183, 179, 156, 163, 156, 187, 156,
        167, 156, 174, 170, 183, 179, 174, 179, 170, 159, 187]


mean, l, upper_bound = mean_confidence_interval(data)
print(f"Mean: {mean}")
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")


results = scipy.stats.ttest_1samp(data, popmean=170, alternative='greater')
print("One-sample t-test results:")
print(f"T-statistic: {results.statistic}")
print(f"P-value: {results.pvalue}")
````
